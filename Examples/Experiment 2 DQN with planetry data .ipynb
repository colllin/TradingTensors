# Enhancing State Space with Planetry Data?

In this experiment, lets explore how the position of Moon affect our Agent's behaviour. Planet Calculation is made based datetime values using Swiss Ephemeris method (pyswisseph). (This experiment is purely for fun as there has been no proven evidence in the predictive power of this additional information)

```python
from tradingtensors.Environments.CustomEnvs import OandaEnv
from tradingtensors.Agents.Q_learning import DQNAgent

#Define Environment
env = OandaEnv(
    INSTRUMENT = "EUR_USD", 
    TIMEFRAME = "H1", 
    lookback_period= 3,
    trade_duration=3,
    planet_data={"Moon" : ["right asc.", "declination", "distance"]}, # Require these coordinates of Moon
    PLANET_FORWARD_PERIOD=2, #Look ahead 2 periods of Moon Coordinates 

)

#Define Agent - DQN

PATH = r'Enter destination files'

agent = DQNAgent(
    env= env,
    PARENT_PATH = PATH #Path to save your models
)
```

```python
agent.train(
CONVERGENCE_THRESHOLD=700 #Setting a more stringent convergence criteria than before
) #use Default Settings
```

```python
agent.trainSummary()
```

```python
'''Surprisingly, model is able to converge even with a more stringent criteria. Maybe pure luck?'''
agent.episodeReview(44)
```

```python
''' Test Case does not do that badly either'''
agent.test(44)
```
